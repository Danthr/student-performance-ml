```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/student-data.csv")
df.head()
```

# Target Variable Analysis

```python
plt.figure()
sns.histplot(df['G3'], bins=20)
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade')
plt.ylabel('Count')
plt.show()
```

Distribution is skewed.
Less low performers more average performers

# Key Numerical Features vs G3

```python
plt.figure()
sns.boxplot(x='studytime', y='G3', data=df)
plt.title('Final Grades vs Study Time')
plt.show()
```

More study time = Higher Grades

```python
plt.figure()
sns.scatterplot(x='absences', y='G3', data=df)
plt.title('Final Grades vs Absences')
plt.show()
```

Less absences lead to more grades

# Categorical Features vs G3

```python
plt.figure()
sns.boxplot(x='internet', y='G3', data=df)
plt.title("Internet Access vs Final Grades")
plt.show
```

# Correlation Analysis

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
